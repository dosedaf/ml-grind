```python
import cv2
import matplotlib.pyplot as plt

# Read color image
image = cv2.imread('imgs/car.jpg')
# plt.imshow(image, cmap=plt.cm.binary)
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
normalized_gray_image = cv2.normalize(
    gray_image, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX
)

normalized_color_image = cv2.cvtColor(
    normalized_gray_image, cv2.COLOR_GRAY2BGR
)

imgs = [image, gray_image, normalized_gray_image, normalized_color_image, rgb_image]

for i in range(5):
    plt.subplot(3, 2, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(imgs[i], cmap=plt.cm.binary)
```
